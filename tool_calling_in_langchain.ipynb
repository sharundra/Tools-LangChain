```python
from dotenv import load_dotenv
load_dotenv()
```

```python
from langchain_openai import ChatOpenAI
from langchain_core.tools import tool
from langchain_core.messages import HumanMessage
```

```python
@tool
def mult_and_sqrt(x:int, y:int) -> float:
    """
    this function multiplies two numbers and then takes the square root of the product
    """
    import math
    return math.sqrt(x * y)
```

```python
llm = ChatOpenAI()
llm_with_tools = llm.bind_tools([mult_and_sqrt])
```

```python
result = llm_with_tools.invoke("Hi, How are you")
print(result)
```

```python
result = llm_with_tools.invoke("What is the square root of 4 multiplied by 3? How did you solve my question")
print(result)
print(result.tool_calls)
print(result.tool_calls[0])
```

```python
query = HumanMessage('What is the square root of 4 multiplied by 3?')
messages = [query]
```

```python
result = llm_with_tools.invoke(messages)
messages.append(result)
```

```python
mult_result = mult_and_sqrt.invoke(result.tool_calls[0])
messages.append(mult_result)
```

```python
final_result = llm_with_tools.invoke(messages)
print(final_result.content)
```
